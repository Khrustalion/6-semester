```python
import cv2

import numpy as np
import matplotlib.pyplot as plt
```

```python
def draw_rectangle(img, pt1, pt2, color=(255, 0, 0), thickness=2, fill=False):
    out = img.copy()
    H, W = out.shape[:2]

    x1, y1 = pt1
    x2, y2 = pt2

    x1, x2 = sorted((int(x1), int(x2)))
    y1, y2 = sorted((int(y1), int(y2)))

    x1 = max(0, min(W - 1, x1))
    x2 = max(0, min(W - 1, x2))
    y1 = max(0, min(H - 1, y1))
    y2 = max(0, min(H - 1, y2))

    color = np.array(color, dtype=out.dtype)

    if fill:
        out[y1:y2+1, x1:x2+1] = color
        return out

    t = max(1, int(thickness))

    # Верхняя полоса
    out[y1:y1+t, x1:x2+1] = color
    # Нижняя полоса
    out[y2-t+1:y2+1, x1:x2+1] = color
    # Левая полоса
    out[y1:y2+1, x1:x1+t] = color
    # Правая полоса
    out[y1:y2+1, x2-t+1:x2+1] = color

    return out
```
